# Looking for BP using the $\hat{Z}_1 \hat{Z}_2$ observable

We start by studying a simple case: the minimization of an expectation value of a simple, two qubits, observable. In order to look for BP, we will study the cocentration of the cost function and its gradient using different number of qubits.

In order to do that, we calculate the variances of the observable and its derivatives. If there is a BP, one should expect to see an exponential reduction of the variances as the number of qubits increase.

First, we import our custom modules that includes functions capable of expand the observable, build different ansatzs and calculate the variances for different numbers of qubits.

```python
import sys
sys.path.append('../../')

from src import customFunc as cf, ansatzs as anz

from qiskit.quantum_info import SparsePauliOp
```

We define the observable:

```python
observable = SparsePauliOp.from_list([("ZZ",1)])
print(observable)
```

We use our custom deep ansatz:

```python
# set the number of layer per qubit
def ansatz_fun(num_qubits):
    return anz.build_deep_ansatz(num_qubits, 10)
```

Calculate the variances concentration:

```python
data = cf.variance_vs_nQubits(ansatz_fun, 2, 15, observable, 0, 100, print_info=True, plot_info=True, do_regress=True)
```

We can see a clearly decreasing linear tendecy. Taking into account that the vertical axis is in logaritmic scale, we conclude that we have found an exponential concentration of the variances, thus, a barren plateau.
